fig, ax = plt.subplots()

These two variables now hold the two core objects used for all types of plotting operations. First object fig, short for figure, imagine it as the frame of your plot. You can resize, reshape the frame but you cannot draw on it. On a single notebook or a script, you can have multiple figures. Each figure can have multiple subplots. Here, subplot is synonymous with axes. The second object, ax, short for axes, is the canvas you draw on. Or rephrasing, it is the blank sheet you can plot and hold your data. An axes object can only belong to one figure.

```python
fig, ax = plt.subplots(ncols=2, figsize=(6, 4))
ax[0].plot(x, y, label="Line 1")
ax[0].set_xlabel("X-axis")
ax[0].set_ylabel("Y-axis")
ax[0].legend()
ax[1].plot(y, x, label="Line 2")
ax[1].set_xlabel("X-axis")
ax[1].set_ylabel("Y-axis")
ax[1].legend()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 4))
ax[0,0].plot(x, y, label="Line 1")
ax[0,0].set_xlabel("X-axis")
ax[0,0].set_ylabel("Y-axis")
ax[0,0].legend()
ax[1,1].plot(y, x, label="Line 2")
ax[1,1].set_xlabel("X-axis")
ax[1,1].set_ylabel("Y-axis")
ax[1,1].legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 4))

# fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label="Line 1")
ax.plot(y, x, label="Line 2")
# Customizing axes
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_xticks([1, 2, 3, 4, 5], ["A", "B", "C", "D", "E"])
ax.set_yticks([1, 5, 10], ["Low", "Mid", "High"])
ax.grid(True)

# Adding legends

ax.legend()

# Adding annotations
ax.annotate("Important Point", xy=(2, 5), xytext=(3, 6),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('housing-updated.csv')
data.head()
```

```python
plt.scatter(data['area'], data['price'])
plt.xlabel('area')
plt.show()
```

```python
to_plot_columns = ['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'furnishingstatus']
```

```python
len(to_plot_columns)
```

```python
list(zip([1,2], ['a', 'b']))
```

```python
fig, axs = plt.subplots(nrows=4, ncols=3, figsize=(15, 12))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("Housing prices vs features")
for column, ax in zip(to_plot_columns, axs.ravel()):
    # filter df for ticker and plot on specified axes
    ax.scatter(data[column], data['price'])
    
    # chart formatting
    ax.set_title(column)
    ax.set_xlabel(column)
    ax.set_ylabel('price')
plt.show()
```

```python
bedrooms_price_list = []
for bedrooms_counts in range(1,7):
    temp = data.loc[data['bedrooms'] == bedrooms_counts, 'price'].values
    bedrooms_price_list.append(temp)
```

```python
plt.boxplot(bedrooms_price_list)
plt.show()
```

```python
data.boxplot(column = 'price', by = 'bedrooms')
plt.show()
```

```python
categorical_columns = ['bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'furnishingstatus']
```

```python
len(categorical_columns)
```

```python
fig, axs = plt.subplots(nrows=4, ncols=3, figsize=(15, 12))
plt.subplots_adjust(hspace=0.5)
fig.suptitle("Housing prices vs categorical features")
for column, ax in zip(categorical_columns, axs.ravel()):
    # filter df for ticker and plot on specified axes
    data.boxplot(column = 'price', by = column, ax=ax)
plt.show()
```
