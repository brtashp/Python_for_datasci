### Question 1:
Write a function named "data_preprocessing" that accepts the dataset in dataframe as only argument and does the following tasks and returns the updated dataframe:<br/>
1. Drop "PassengerId" and "Ticket" columns from the dataframe.(1 pts)
2. Check if there are any rows that have missing entries for any of the columns and remove such rows from the given dataset.(1 pts)
3.  Convert the data type of "Fare" and "Age" to int.(1 pts)
4. Rename "SibSp" to "Siblings/Spouses Aboard", and "Parch" to "Parents/Children Aboard".(1 pts)

```python
#Work on your solution here and after. 
#make sure to follow the given structure below

copy_titanic = titanic.copy() 
# makes a copy so the original dataframe does not get messed up 

#Write your function here
def data_preprocessing(copy_titanic):

    copy_titanic = copy_titanic.drop(["PassengerId", "Ticket"], axis=1)
    # removes the PassengerID and ticket columns (axis=1 is for columns, axis=0 is for rows)
    
    copy_titanic = copy_titanic.dropna()
    # removes the associated rows (rows that were for passangerid and ticket now have a NaN value, the dropna() function drops 
    # rows with the NaN value) 
    
    copy_titanic["Fare"] = copy_titanic["Fare"].astype(int)
    copy_titanic["Age"] = copy_titanic["Age"].astype(int)
    # converts "Fare" and "Age" to int
    
    copy_titanic = copy_titanic.rename(columns={"SibSp": "Siblings/Spouses Aboard", "Parch": "Parents/Children Aboard"})
    # renames the columns columns
    
    return copy_titanic
#Call your function here
new_titanic = data_preprocessing(copy_titanic)

#Report results
new_titanic.head()
```

## Question 2:
We are interested in adding new columns which are so-called features.<br/>
Create a new function called "feature_extraction" which does the following updates to the existing dataframe.<br/>

1. Convert "Embarked" column into numeric. Replace "S" with 0; replace "C" with 1 and replace "Q" with 2. For the missing values under "Embarked", just fill with the most common one (ex: if there are 100 0, 200 1, and 300 2 in column, then 2 is the most common. fill missing values with 2). (2 pts)
2. The column named "Cabin" includes strings with letters and numbers, for example "C85" and the letter refers to the cabin group. This column should indicate integer value of 1 for cabin name starting with letter "A". The rest of the numbers should appear according to this dictionary: {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}. Any missing and/or other character entries should be replaced with 0. If there are multiple names, only the first character matters (ex: "F C100 B90 30", then it is F and "F":6) (2 pts)
3. (2 pts) Any missing values from "Fare" column should be replaced with 0, and update it according to the following conditions:<br/>
  if 0 <= 'Fare' < 10, then 'Fare' = 0<br/>
  if 10 <= 'Fare' < 20, then 'Fare' = 1<br/>
  if 20 <= 'Fare' < 30, then 'Fare' = 2<br/>
  if 30 <= 'Fare' < 100, then 'Fare' = 3<br/>
  if 100 <= 'Fare' < 200, then 'Fare' = 4<br/>
  if 200 <= 'Fare' then 'Fare' = 5<br/>

**Note:** Use the orignal datset without considering the modifications from **Question 1**


```python
#Work on your solution here and after. 
#Write your function here
copy_titanic_two = titanic.copy()
# Perform feature extraction
copy_titanic_two = feature_extraction(copy_titanic_two)

# Display the updated DataFrame
copy_titanic_two.head()

def feature_extraction(copy_titanic_two):
    
    copy_titanic_two["Embarked"] = copy_titanic_two["Embarked"].map({"S": 0, "C": 1, "Q": 2})
    # line maps the values in embarked to numeric values (S to 0, C to 1 and Q to 2)
    most_common_embarked = copy_titanic_two["Embarked"].mode().iat[0]
    # finds the mode (most common) value using the mode function, creates mode variable 
    # the .iat[] was used because I kept getting an error since there were multiple values
    # from using the mode function, I found that function lets me just take the first value 
    # of all the same values (the first element)
    copy_titanic_two["Embarked"] = copy_titanic_two["Embarked"].fillna(most_common_embarked)
    # fills in any missing values in the embarked column with the most common value
    
    copy_titanic_two["Cabin"] = copy_titanic_two["Cabin"].str.extract(r'([A-Za-z])')
    # extracts the first letter from the values in the cabin column
    cabin_dict = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}
    # creates a dictionary that maps the alphabetic values to an integer 
    copy_titanic_two["Cabin"] = copy_titanic_two["Cabin"].map(cabin_dict).fillna(0)
    # aligns alphabetic values in cabin to corresponding integer values and fills in 
    # any missing values with 0
    # .map function, defines the correspondence between original values and their replacements

    copy_titanic_two["Fare"] = copy_titanic_two["Fare"].fillna(0)
    # fills in missing values with 0 
    
    copy_titanic_two.loc[(copy_titanic_two["Fare"] >= 0) & (copy_titanic_two["Fare"] < 10), "Fare"] = 0
    copy_titanic_two.loc[(copy_titanic_two["Fare"] >= 10) & (copy_titanic_two["Fare"] < 20), "Fare"] = 1
    copy_titanic_two.loc[(copy_titanic_two["Fare"] >= 20) & (copy_titanic_two["Fare"] < 30), "Fare"] = 2
    copy_titanic_two.loc[(copy_titanic_two["Fare"] >= 30) & (copy_titanic_two["Fare"] < 100), "Fare"] = 3
    copy_titanic_two.loc[(copy_titanic_two["Fare"] >= 100) & (copy_titanic_two["Fare"] < 200), "Fare"] = 4
    copy_titanic_two.loc[copy_titanic_two["Fare"] >= 200, "Fare"] = 5
    # updates values in Fare based on conditions, assigns a range of (0,5) for each of the 6 categories
    
    return copy_titanic_two

#Call your function here
new_titanic = feature_extraction(new_titanic)

#Report results
new_titanic.head()
```
